```python
#vals=list(stocks.values())
vals= [f'{k} ({v})' for k,v in stocks.items()]
print(vals)
fig = make_subplots(rows=1, cols=len(stocks.keys()), subplot_titles=vals)
i=1
for s in stocks.keys():
    vals = yf.download(s, period='MAX')
    fig.add_trace(
        go.Scatter(x=vals.index, y=vals['Close']),
        row=1, col=i
    )
    i=i+1

fig.update_layout(showlegend=False, height=600, width=2000)
fig.show()
```
