```python
import xml.etree.ElementTree as ET
import pandas as pd
import datetime 
import matplotlib.pyplot as plt
import seaborn as sns
import os as os 
import plotly.express as px
```

```python
withings_data = '/Users/andrewspruce/Downloads/Withings.zip/weight.csv'
df = pd.read_csv(withings_data)
df = df.fillna(1)

# Using pandas.to_datetime() to convert pandas column to DateTime
df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d %H:%M:%S")
df = df[df['Date'] > '01-Oct-2021']
```

```python
fig = px.scatter(df, 
                x='Date', 
                y='Weight (kg)',
                size='Fat mass (kg)',
                trendline="lowess", trendline_options=dict(frac=0.1) )
fig.add_hline(y=84,line_color='red')
fig.update_layout(yaxis_range=[75,100])
fig
```
